```
import nltk
nltk.download('all')
```

```
from nltk.corpus import state_union
from nltk.tag import pos_tag
from nltk.tokenize import PunktSentenceTokenizer,word_tokenize
```

```
train_text = state_union.raw("2005-GWBush.txt")
sample_text = state_union.raw("2006-GWBush.txt")
```

```
custom_sentence_tokenizer = PunktSentenceTokenizer(train_text)
tokenized_sentence = custom_sentence_tokenizer.tokenize(sample_text)
```

```
def content():
    try:
      for t in tokenized_sentence:
        words = word_tokenize(t)
        tagged = pos_tag(words)
        chunkGram = r""" Chunk: {<NN>+?} """
        newChunk = r""" newChunk: {<.*>+} 
                                }<VB.?|IN|DT>+{"""

        chunkParser = nltk.RegexpParser(chunkGram)
        newChunkParser = nltk.RegexpParser(newChunk)
        newChunked = newChunkParser.parse(tagged)
        chunked = chunkParser.parse(tagged)

        print(chunked)

    
    except Exception as e:
        print(str(e))

content()
   
```
